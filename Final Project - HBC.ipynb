```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("datasets/ks-projects-201801.csv")

df.head()
```

```python
df.count()
```

```python
newdf = df.dropna()
newdf.count()
```

```python
newdf = newdf.drop(['usd_pledged_real', 'usd_goal_real'], axis=1)
newdf.head()
```

```python
newdf['pledged'] = newdf['pledged'].apply(lambda x: int(x))
newdf['goal'] = newdf['goal'].apply(lambda x: int(x))
newdf.head()
```

```python
def togoal(x, y):
    return (x / y)*100

newdf['completion'] = togoal(newdf['pledged'], newdf['goal']).apply(lambda x: round(x, 2))
newdf.head()
```

```python
newdf['deadline'] = pd.to_datetime(newdf['deadline'])
newdf['launched'] = pd.to_datetime(newdf['launched'])

def tlen(x, y):
    return x - y

newdf['campaign_length'] = tlen(newdf['deadline'], newdf['launched'])
newdf.head()
```

```python
us_df = newdf.loc[newdf['country'] == 'US'] 
us_df['state'] = us_df['state'].map({'failed':0, 'successful':1})
us_df = us_df.dropna()
us_df['state'] = us_df['state'].apply(lambda x: int(x))
us_df.head()
```

```python
us_df2 = us_df[['ID','state', 'completion', 'campaign_length', 'goal']]
us_df2 = us_df2.set_index('ID')
us_df2['campaign_length'] = us_df2['campaign_length'].dt.days
us_df2.head()
```

```python
#pretty weak correlations, strongest is neg correlation btwn campaign length & state
us_df2.corr()
```

```python
us_df.to_csv('US Kickstarter Data.csv', index=False)
```
